```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Household_data.csv')  # Replace 'your_dataset.csv' with the actual filename

# Calculate the total population across all areas
total_population = df['total_Person'].sum()

# Calculate the distribution of each household size
household_sizes = ['Single_Person', 'Two_Person', 'Three_Person', 'Four_Person', 'Five_Person', 'Six_Person']
size_distribution = [df[size].sum() for size in household_sizes]

# Plot the total population and household size distribution
plt.figure(figsize=(10, 6))
plt.bar(['Total Population'] + household_sizes, [total_population] + size_distribution, alpha=0.7)

plt.xlabel('Household Size')
plt.ylabel('Population')
plt.title('Total Population and Household Size Distribution')
plt.tight_layout()

# Show the plot
plt.savefig('householdbar.pdf')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Household_data.csv')  # Replace 'your_dataset.csv' with the actual filename

# Calculate the total population for each household size
total_population = df[['Single_Person', 'Two_Person', 'Three_Person', 'Four_Person', 'Five_Person', 'Six_Person']].sum()

# Plot the pie chart
labels = ['Single Person', 'Two Person', 'Three Person', 'Four Person', 'Five Person', 'Six Person']
plt.figure(figsize=(8, 8))
plt.pie(total_population, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Population Distribution by Household Size')

# Show the plot
plt.savefig('householdpie.pdf')
plt.show()
```
